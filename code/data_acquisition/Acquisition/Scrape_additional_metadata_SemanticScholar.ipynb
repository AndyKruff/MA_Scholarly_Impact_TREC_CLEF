```python
import json
import requests
```

```python
# Insert your API key here
api_key = ""

headers = {
    "x-api-key": api_key  # Setze den x-api-key als Header
}

params = {
        "limit": 1,
        "fields": "corpusId,externalIds,citationCount,influentialCitationCount,authors,journal,title,citations,venue,publicationVenue,year,referenceCount,isOpenAccess,openAccessPdf,fieldsOfStudy,s2FieldsOfStudy,publicationTypes"
    }
```

```python
# Add all necessary metadata regarding citations 

params_citations = ["paperId","corpusId","title","venue","publicationVenue","year","authors", "externalIds", "referenceCount","citationCount", "influentialCitationCount", "isOpenAccess","openAccessPdf","fieldsOfStudy","s2FieldsOfStudy","publicationTypes","journal"]
string = ""
for i in params_citations:
    string += ",citations." + i

params["fields"] += string
```

## Additional metadata for TREC

```python
# Read in original matadata file for Semantic Scholar (TREC)
with open('../../../data/SemanticScholar_TREC.json', 'r', encoding="utf-8") as file:
    Semantic_Trec = json.load(file)
```

```python
import time
more_infos_dict = {}

# Request the TREC documents by Semantic Scholar IDs and request the additional metadata

for i in Semantic_Trec:

    try:
        url = f'https://api.semanticscholar.org/graph/v1/paper/{Semantic_Trec[i]["paperId"]}'
        resp = requests.get(url, params=params, headers=headers).json()
        more_infos_dict[i] = resp
        time.sleep(4)  
    except:
        print(i)
```

```python
with open('../../../data/SemanticScholar_TREC_additional_metadata.json', 'w', encoding="utf-8") as f:
    json.dump(more_infos_dict, f, ensure_ascii=False, indent=4)
```
