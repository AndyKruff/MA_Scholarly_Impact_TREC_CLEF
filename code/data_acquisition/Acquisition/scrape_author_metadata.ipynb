```python
import json
```

```python
# Read in all metadata for all three proceedings
with open("../../../data/OpenAlex_LNCS.json", 'r', encoding="utf-8") as file:
        OpenAlexCitations_lncs = json.load(file)
with open("../../../data/OpenAlex_TREC.json", 'r', encoding="utf-8") as file:
        OpenAlexCitations_trec = json.load(file)
with open("../../../data/OpenAlex_CEUR.json", 'r', encoding="utf-8") as file:
        OpenAlexCitations_ceur = json.load(file)
```

```python
# Create a list were all authorship IDs from OpenAlex are stored for all proceedings

authorship_list_ids = []

for citation in OpenAlexCitations_lncs.values():
    for authorship in citation["authorships"]:
        author_id = authorship["author"].get("id")
        if author_id:  
            authorship_list_ids.append(author_id)  
            
# Additionally add the TREC Author IDs from OpenAlex
for citation in OpenAlexCitations_trec.values():
    for authorship in citation["authorships"]:
        author_id = authorship["author"].get("id")
        if author_id:  
            authorship_list_ids.append(author_id)  

# Additionally add the CEUR Author IDs from OpenAlex
for citation in OpenAlexCitations_ceur.values():
    for authorship in citation["authorships"]:
        author_id = authorship["author"].get("id")
        if author_id:  
            authorship_list_ids.append(author_id)  
```

```python
# Remove potential duplicated author IDs
authorship_list_ids = list(set(authorship_list_ids))
```

```python
import requests
author_works_dict = {}
counter = 0

# Request all the works of an author and calculate the amount of citations per year for every author
for i in authorship_list_ids:
    counter += 1
    print(counter, "/" , len(authorship_list_ids))
    url = f"https://api.openalex.org/works?filter=author.id:{i.replace('https://openalex.org/', '')}"

    # Send GET-Request
    response = requests.get(url)
    try:
        data = response.json()["results"]
        metadata = []
        for j in data:
            metadata.append([j["cited_by_count"], j["counts_by_year"], j["id"], j["publication_year"]])
        author_works_dict[i] = metadata
    except:
        author_works_dict[i] = []
```

```python
with open('../../../data/author_metadata_publications.json', 'w', encoding='utf-8') as json_file:
    json.dump(author_works_dict, json_file, ensure_ascii=False, indent=4)
```
