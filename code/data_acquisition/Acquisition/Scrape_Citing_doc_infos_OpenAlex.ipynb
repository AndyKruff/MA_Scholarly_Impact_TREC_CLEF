```python
import json
import requests
from pyalex import Works, Authors, Sources, Institutions, Concepts, Publishers, Funders
```

# Request metadata for Citing Docs of TREC proceeding paper

```python
with open('../../../data/OpenAlex_TREC.json', 'r', encoding="utf-8") as file:
    OpenAlexTrec = json.load(file)
```

```python
citing_doc = {}
for i in OpenAlexTrec:
    response = requests.get(OpenAlexTrec[i]["cited_by_api_url"]).json()
    citing_doc[i] = response["results"]
```

```python
with open('../../../data/OpenAlex_TREC_citing_doc.json', 'w', encoding="utf-8") as f:
    json.dump(citing_doc, f, ensure_ascii=False, indent=4)
```
