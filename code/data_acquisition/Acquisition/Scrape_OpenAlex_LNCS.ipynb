```python
def starts_with_any(doi, strings):
    return any(doi.startswith(s) for s in strings)
```

```python
# Exlude the book papers from the dataframe

df_lncs_filtered = df_lncs[~df_lncs['DOI'].apply(lambda x: starts_with_any(x, doi_list))]
```
