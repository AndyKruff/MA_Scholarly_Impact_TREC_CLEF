```python
# Read in LNCS metadata

df_lncs = pd.read_parquet("../../../data/metadata_LNCS.parquet")
```

```python
# Removing urls linking to the special book releases of CLEF

doi_list = ["https://doi.org/10.1007/978-3-031-04431-1",
"https://doi.org/10.1007/978-3-642-23008-0",
"https://doi.org/10.1007/978-3-030-22948-1",
"https://doi.org/10.1007/978-3-642-15181-1"]
```

```python
def starts_with_any(doi, strings):
    return any(doi.startswith(s) for s in strings)
```

```python
# Remove metadata rows related to the books above

df_lncs = df_lncs[~df_lncs['DOI'].apply(lambda x: starts_with_any(x, doi_list))]
```

```python
# Remove titles that are not relevant for the analysis

titles_to_remove = ["Introduction","Introduction","Front Matter","CLEF 2003 Methodology and Metrics", "Analysis of the Reliability of the Multilingual Topic Set for the Cross Language Evaluation Forum", "Evaluation of Information Access Technologies at the NTCIR Workshop","What Happened in CLEF 2004?", "What Happened in CLEF 2005","What Happened in CLEF 2006","Scientific Data of an Evaluation Campaign: Do We Properly Deal with Them?",
                   "What Happened in CLEF 2007", "What Happened in CLEF 2008","What Happened in CLEF 2009","What Happened in CLEF 2009"]
```

```python
df_lncs = df_lncs[~df_lncs['Title'].isin(titles_to_remove)]
```

```python
# Remove duplicated rows 

df_lncs = df_lncs[df_lncs.duplicated(subset=['Title'], keep=False)]
```

```python
# Remove the documents from 2023 

df_lncs = df_lncs[df_lncs["PubYear"] != "2023"]
```

```python
# Save the filtered metadata for LNCS

df_lncs.to_parquet("../../../data/metadata_LNCS.parquet")
```
