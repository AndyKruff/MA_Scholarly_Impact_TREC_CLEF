```python
import os
import re
import time
from grobid_client.grobid_client import GrobidClient
```

# Transformation of CEUR PDFs into XML

```python
# Path to the directory containing the PDF data
path = "../../../data/PDF_data/CEUR_Proceedings"

# Get a list of all directories in the specified path
directories = [x[0] for x in os.walk(path)]
```

```python
directories
```

```python
# Initialize GrobidClient with the specified configuration file
client = GrobidClient(config_path="../../config.json")

# Loop through each directory, starting from the second one
for i in directories[1:]:
    # Search for the volume number in the directory name (e.g., Vol-1234)
    volume = re.search(r"Vol-\d+", i)
    
    # Process the full text document and output the result to the specified directory
    client.process("processFulltextDocument", i,  output=f"../../resources/XML_CEUR/{volume.group(0)}/" ,consolidate_citations=False, tei_coordinates=True, force=True ,  n=10)
```
